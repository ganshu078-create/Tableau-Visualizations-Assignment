#Tableau Visualizations Assignment


# Task 1: Category Performance Analysis

## Business Need
Management wants to know which product category generates the highest revenue.

## Visualization Used
Bar Chart – Total Sales by Category

### Fields Used
- Category → Rows
- Sales → Columns (SUM)

## Findings

| Category | Total Sales |
|-----------|------------:|
| Technology | ₹364,276.41 |
| Furniture | ₹353,637.09 |
| Office Supplies | ₹329,260.88 |

### Insight

The **Technology** category generated the highest sales revenue, followed by Furniture and Office Supplies.

## Business Recommendation

The company should continue investing in Technology products through targeted promotions, inventory planning, and product expansion since this category contributes the highest revenue.

## Conclusion

Technology is the top-performing category and represents the strongest revenue opportunity for the business.


# Task 2: Sales Trend Analysis

## Business Need
Management wants to understand how sales fluctuate over time.

## Visualization Used
Line Chart – Monthly Sales Trend

### Fields Used
- Order Date (Month) → Columns
- Sales → Rows (SUM)

## Findings

The sales trend shows fluctuations throughout the year, indicating seasonal demand patterns and varying customer purchasing behavior.

### Insight

Some months generated significantly higher sales than others, suggesting that customer demand changes over time.

## Business Recommendation

Management can plan promotions, inventory stocking, and marketing campaigns around high-demand periods while improving strategies during slower months.

## Conclusion

Monthly trend analysis helps forecast demand and supports better operational planning.


# Task 3: Regional Contribution Analysis

## Business Need
Understand which region contributes the most to sales.

## Visualization Used
Bar Chart or Pie Chart – Sales by Region

### Fields Used
- Region → Rows
- Sales → Columns (SUM)

## Findings

| Region | Total Sales |
|---------|------------:|
| West | ₹287,787.07 |
| Central | ₹266,483.65 |
| South | ₹250,878.20 |
| East | ₹242,025.46 |

### Insight

The **West Region** contributes the highest sales, while the **East Region** contributes the least.

## Business Recommendation

Management should study the success factors in the West Region and apply similar strategies in lower-performing regions.

## Conclusion

The West Region is the strongest market and can serve as a benchmark for regional growth initiatives.


# Task 4: Profitability Investigation

## Business Need
Identify whether high sales always lead to high profits.

## Visualization Used
Scatter Plot – Sales vs Profit

### Fields Used
- Sales
- Profit

## Findings

The dataset contains several orders with high sales but negative profits.

### Insight

A total of **99 orders** generated losses despite generating revenue.

Examples:

| Order ID | Sales | Profit |
|-----------|--------:|--------:|
| 1068 | ₹4,218.15 | -₹497.94 |
| 1026 | ₹1,102.27 | -₹497.76 |
| 1113 | ₹2,191.27 | -₹497.54 |

## Business Recommendation

Management should investigate pricing strategies, discounts, shipping costs, and procurement expenses for loss-making orders.

## Conclusion

Higher sales do not always guarantee higher profits. Profitability should be monitored alongside revenue.

# Q9. Chart Selection Justification

## Answer

Different chart types were selected because each business question requires a different method of analysis.

| Business Question | Chart Used | Reason |
|------------------|------------|---------|
| Compare Category Sales | Bar Chart | Best for comparing categories |
| Analyze Sales Trend | Line Chart | Shows changes over time |
| Regional Performance | Pie/Bar Chart | Displays contribution by region |
| Sales vs Profit | Scatter Plot | Shows relationship between variables |
| Growth Analysis | Running Total Line Chart | Tracks cumulative growth |
| Category vs Region | Stacked Bar Chart | Compares multiple dimensions |

### Conclusion

Selecting the right visualization improves understanding, highlights patterns quickly, and supports better decision-making.
